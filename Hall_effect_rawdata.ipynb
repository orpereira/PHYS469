```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

def plot_voltage_temp(dataset, title = ''):
    fig, axs = plt.subplots(4, 2, figsize=(8, 8), dpi=200)
    fig.suptitle(title)

    for i in range(1, 8):
        ax = axs[(i-1)//2, (i-1)%2]
        ax.scatter(1/dataset['T'], dataset[f'v{i}'], s=5)
        ax.set_title(f'v{i}')

    fig.supxlabel('1/T (1/K)')
    fig.supylabel('Voltage (V)')
    plt.tight_layout()
    plt.show()
```

## 1. No Heating or Magnetic field

```python
# load csv using pandas
onemA_no_magnet = pd.read_csv('1mA_no_magnet_3.csv', header=2)
plot_voltage_temp(onemA_no_magnet, 'No heating or magnetic field, 1mA')

threemA_no_magnet = pd.read_csv('3mA_no_magnet.csv', header=2)
plot_voltage_temp(threemA_no_magnet, 'No heating or magnetic field, 3mA')

tenmA_no_magnet = pd.read_csv('10mA_no_magnet.csv', header=2)
plot_voltage_temp(tenmA_no_magnet, 'No heating or magnetic field, 10mA')

thirtymA_no_magnet = pd.read_csv('30mA_no_magnet.csv', header=2)
plot_voltage_temp(thirtymA_no_magnet, 'No heating or magnetic field, 30mA')
```

## 2. Heating, no magnetic field

```python
sixV_heating = pd.read_csv('10mA_6V_heating.csv', header=2)
plot_voltage_temp(sixV_heating, '6V heating, 10mA')

twelveV_heating = pd.read_csv('10mA_12V_heating.csv', header=2)
plot_voltage_temp(twelveV_heating, '12V heating, 10mA')

eighteenV_heating = pd.read_csv('10mA_18V_heating.csv', header=2)
plot_voltage_temp(eighteenV_heating, '18V heating, 10mA')
```
